```python
import torch
import librosa
from scipy.io import wavfile
from IPython.display import Audio
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
# loading the pre-trained tokenizer and model
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
file_name = '/content/drive/MyDrive/Colab Notebooks/Speech to Text/sample_audio.wav'
```

```python
Audio(file_name)
```

```python
audio = wavfile.read(file_name)

frame_rate = audio[0]
print(frame_rate)
```

```python
input_audio,srate = librosa.load(file_name, sr=16000)
```

```python
# tokenizing our digitalised audio file
input_values = tokenizer(input_audio, return_tensors="pt").input_values

# un-normalised predictions
logits = model(input_values).logits

pred = torch.argmax(logits,dim=-1)
print(len(pred))
pred
```

```python
# now lets decode this labels into respective texts

output = tokenizer.batch_decode(pred)[0]
```
